```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname = "C:\Windows\Fonts\malgun.ttf").get_name()
rc('font', family=font_name)
mpl.rcParams['axes.unicode_minus'] = False
%matplotlib inline

import FinanceDataReader as fdr
```

```python
dividend = pd.read_csv('dividend.csv', encoding = 'euc-kr')
dividend = pd.DataFrame({'Ratio': dividend.iloc[:,0], 'Name': dividend.iloc[:,1], 'Symbol': dividend.iloc[:,2]})

b = []
for i in dividend['Symbol'] :
    k = str(i).zfill(6)
    b.append(k)
dividend['Symbol'] = b

dividend.Ratio.unique()
```

```python
a = dividend[dividend['Ratio'] == '6~8']
```

```python
plt.figure(figsize=(20,20))

mpl.rc('font', family='nanumgothic')
mpl.rc('axes', unicode_minus=False)

for i in range(len(a)) :
    plt.subplot(5,4,i+1)
    plt.title(a['Name'].iloc[i])
    plt.xlabel('Date')
    plt.ylabel('Price')
    fdr.DataReader(a['Symbol'].iloc[i], '2018-12-20', '2019-01-10')['Close'].plot()

plt.style.use('seaborn')
plt.tight_layout()
plt.show()
```
