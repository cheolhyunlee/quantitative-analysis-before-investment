```python
import FinanceDataReader as fdr
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
%matplotlib inline
```

```python
df_kospi = fdr.DataReader('KS11','2014')
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(18,6))
plt.plot(df_kospi['Close'])
plt.show()
```

```python
df_exchange = fdr.DataReader('USD/KRW','2014')
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(18,6))
plt.plot(df_exchange['Close'])
plt.show()
```

```python
plt.style.use('ggplot')
fig, ax0 = plt.subplots(figsize=(18,6))
ax1 = ax0.twinx()

ax0.set_title('Inverse relationship between index and exchange rate 2014 ~ ', fontsize = 18)
ax0.plot(df_kospi['Close'], c = 'blue', label='KOSPI')
ax0.set_ylabel("Point", fontsize = 15)
ax0.grid(True)

ax1.plot(df_exchange['Close'], c = 'magenta', label = 'USD/KRW')
ax1.set_ylabel("KRW", fontsize = 15)

fig.legend()

ax0.set_xlabel('Time', fontsize = 15)
plt.show()
```
